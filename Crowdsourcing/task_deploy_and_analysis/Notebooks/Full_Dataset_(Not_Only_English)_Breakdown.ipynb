```python
to_remove = df2.columns.tolist()
to_remove.remove('claim_id')
df_crowd_r = df_crowd.drop(to_remove, axis=1)
df_crowd_r.head()
```

```python
df_crowd_r['claim_id_2'] = df_crowd_r['claim_id']
df_j = df2.join(df_crowd_r.set_index('claim_id'), on = 'claim_id', how='outer', lsuffix='_caller', rsuffix='_other')
df_j.head(100)
```

```python
for i, row in df_j.iterrows():
    if not row['claim_id_2'] is np.nan:
        try:
            assert row['claim_id'] == row['claim_id_2']
        except Exception:
            print(row['claim_id'], row['claim_id_2'])
            continue
```

```python
df_j.to_csv('./data/campaign/campaign_sampled_df_verbalised_unk_fixed_crowd_merged_FULL.csv',index=None)
```

```python
import json
import ast
json_data = []

for i, row in df_j.iterrows():
    try:
        json_data_row = {
            'claim_id': row['claim_id'],
            'rank': row['rank'],
            'subject_id': row['entity_id'],
            'property_id': row['property_id'],
            'subject_label': row['entity_label'],
            'property_label': row['property_label'],
            'object_label': row['object_label'],
            'subject_dec': row['entity_desc'],
            'property_desc': row['property_desc'],
            'object_desc': row['object_desc'],
            'subject_alias': row['entity_alias'] if row['entity_alias'] == 'no-alias' else ast.literal_eval(row['entity_alias']),
            'property_alias': row['property_alias'] if row['property_alias'] == 'no-alias' else ast.literal_eval(row['property_alias']),
            'object_alias': row['object_alias'] if row['object_alias'] == 'no-alias' else ast.literal_eval(row['object_alias']),
            'object_datatype': row['datatype'],
            'object': ast.literal_eval(row['datavalue']),
            'theme_root_class_id': row['theme_entity_id'],
            'theme_label': row['theme_entity_label'],
            'verbalisation': row['verbalisation'],
            'verbalisation_unk_replaced': row['processed_verbalisation'],
            'sampling_weight': row['sampling_weight']
        }
        if row['claim_id_2'] is not np.nan:
            json_data_row.update(
            {'annotations' : {
                    'fluency_scores': ast.literal_eval(row['fluency']),
                    'fluency_mean': row['fluency_mean'],
                    'fluency_median': row['fluency_median'],
                    'adequacy_scores': ast.literal_eval(row['adequacy']),
                    'adequacy_majority_voted': row['adequacy_maj'],
                    'adequacy_percentage': row['adequacy_perc']
            }})
        else:
            json_data_row.update(
                {'annotations' : None}
            )
        json_data.append(json_data_row)
    except Exception:
        print(row)
        raise
```

```python
with open('./data/campaign/WDV_JSON.json', 'w', encoding='utf-8') as f:
    json.dump(json_data,f,indent=2,ensure_ascii=False)
```
