```python
import sqlite3
import json
from collections import Counter
import random
```

```python
to_extract_picks = []
with open('to_extract_picks.txt','r') as f:
    for line in f:
        to_extract_picks.append(line.strip())
```

```python
len(to_extract_picks)
```

```python
extracted_picks = []
with open('extracted_picks.txt','r') as f:
    #print(f.readline())
    for line in f:
        line = line.strip()
        extracted_picks.append(line.split(' - '))
```

```python
len(extracted_picks)
```

```python
extracted_pick_ids = [i[0] for i in extracted_picks]
extracted_pick_ids == to_extract_picks
```

```python
db = sqlite3.connect('wikidata_claims_refs_parsed.db')
cursor = db.cursor()
```

```python
def check_pick(pick):
    entity_id = pick[1]
    cursor.execute('select * from claims where entity_id = ?',[entity_id])
    results = cursor.fetchall()
    if len(results)>0:
        return True
    else:
        return False
```

```python
not_extracted_picks = []
for pick in extracted_picks:
    if not check_pick(pick):
        not_extracted_picks.append(pick)
```
